### 类
同一个类的函数调用 ：  self.fun() 

所有类内的函数定义都要加self， self代表调用者。

同一个类内调用函数的时候都要加上self

```python
class Solution:
    
    def fun1(self,a,b):
        if a> 0 and b> 0 :
            return a+b
        else:
            return self.fun1(a+1,b+1)
    

    def fun2(self,c,d):
        return self.fun1(c,d)

s = Solution()
print(s.fun2(-5,3)) 


```

### 装饰器

```python
def deco(fun):       # 定义装饰器函数，输入类型为一个函数，  输出类型依然是一个函数。
    def wrapper():   #  自定义装饰器一定要使用wrapper函数
        fun()        # 调用传入的函数
        print('a')  #定义新的功能
    return wrapper   #返回一个添加了新功能的函数

@deco
def fun1():
    print('b')
    
@deco
def fun2():
    print('c')


fun2()
```

### 类方法
装饰器 classmethod

@classmethod就是让实例在调用被@classmethod装饰的方法时，隐式的传入模具（类）本身。

一句话总结各种方法的区别：
- 所谓实例方法就是第一个参数是self，
- 所谓类方法就是第一个参数是class，
- 而静态方法不需要额外的参数。



```python
class S: 
    @classmethod
    def cls_name(cls):
        print(cls.__name__)
s = S() 
s.cls_name()
```

```python
class Date:
    def __init__(self,day,month,year):
        self.day = day 
        self.month = month 
        self.year = year 
        
    @classmethod 
    def from_string(cls,date_as_string):
        day , month , year  = map(int,date_as_string.split('-'))
        date = cls(day,month,year)
        return date
    
date = Date.from_string('08-01-2018')
print(date.day,date.month,date.year)
```

```python
class Date:
    def __init__(self,day,month,year):
        self.day = day 
        self.month = month 
        self.year = year 
    def fun(self,a,b,c):
        return Date(a,b,c)
```

### __repr__ 自动打印

```python
class Point:
    def __init__(self,x,y):
        self.x , self.y = x ,y 
    def __repr__(self):
        return 'Point(x=%s,y=%s)'%(self.x , self.y)
```

```python
p = Point(1,2)
p
```

## 继承
- 子类可以继承父类的方法 和 定义自己的方法
- 子类也可以有子类
- 继承的意义是代码的复用

### super（ ）  父类操作
在类的继承中，如果重新定义某个方法，父类的方法就会被覆盖.
有时我们希望同时调用原来的父类功能，这时就要用到super方法。

```python
##正常的 
class Animal:
    def __init__(self,name):
        self.name = name
    def eat(self):
        print('eat')

class Dog(Animal):
    def __init__(self,name,foot):
        self.name = name 
        self.foot = foot
 
    def eat(self):
        print('eating')
        
d = Dog('tim','4')  #初始化时必须传入两个参数

d.name
```

```python
# super
class Animal:
    def __init__(self,name):
        self.name = name
        
    def eat(self):
        print('eat')
        

class Dog(Animal):
    def eat(self):
        super().eat()   # 加上super使得父类的eat函数也会生效
        print('eating')
        
d = Dog('tim')

d.eat()
```
