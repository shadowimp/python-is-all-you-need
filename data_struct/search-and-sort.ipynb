### 1.binary search

````python
def binary_search(nums, target) : 
    left , right = 0 , len(nums)-1 
    #left == right时 就是数组中仅剩一个元素，mid==left==right 其必然被返回    
    while left <= right :
        mid = (left + right)//2 
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1 
        else: 
            right = mid - 1 
    return -1 
````

````python
binary_search( [1,2,4,7,8] , 9)
````

### 2.bubble sort

````python
def bubble_sort(nums):
    for j in range(0,len(nums)-1):
        for i in range(0 ,len(nums)-1-j):
            if nums[i] > nums[i+1]:
                nums[i] , nums[i+1] = nums[i+1] , nums[i]
    return nums
````

````python
bubble_sort( [2,4,6,6,3,1,2])
````

````python
def sum_recur(nums):
    if not nums:
        return 
    if len(nums) == 1 :
        return nums[0]
    return nums[0] + sum_recur(nums[1:])
````

### 4.merge sort

````python
def merge_sort(nums):
    if len(nums) < 2 :
        return nums
    mid = len(nums)//2
    return merge(merge_sort(nums[:mid]),  merge_sort(nums[mid:]) ) 

def merge(nums1 , nums2):
    nums3 = []
    i , j = 0 ,0 
    while  i<len(nums1) and j<len(nums2) :
        if nums1[i] < nums2[j]:
            nums3.append(nums1[i])
            i += 1 
        else:
            nums3.append(nums2[j])
            j += 1 
    if i < len(nums1):
        nums3.extend(nums1[i:])
        return nums3
    if j < len(nums2):
        nums3.extend(nums2[j:])
    return nums3     
````

````python
merge_sort( [2,4,6,6,3,1,2,1,8,2])
````

````python
def quick_sort(nums):
    if len(nums) <2 :
        return nums
    left = [num for num in nums if num<nums[0]]
    right = [num for num in nums if num> nums[0]]
    middle = [num for num in nums if num == nums[0]]
    return quick_sort(left) + middle + quick_sort(right)
````

### 6.select sort

````python
def select_sort(nums):
    for i in range(1,len(nums)):
        max_index = 0
        for j in range(1,len(nums)-i):  
            if nums[j]> nums[max_index] :
                max_index = j 
        nums[max_index] , nums[-i] = nums[-i] , nums[max_index]
    return nums
````

````python
select_sort([2,4,6,1,6,3,2])
````

### 7.Insert sort

````python
def insert_sort(nums):
    for i in range(1,len(nums)):
        j = i
        while j > 0 and nums[j] < nums[j-1]:
            nums[j] , nums[j-1] = nums[j-1] , nums[j]
            j -= 1 
    return nums       
````

````python
insert_sort([2,4,6,1,6,3,2])
````

### 8 heap sort
```
堆排序的方法如下，把最大堆堆顶的最大数取出，将剩余的堆继续调整为最大堆，再次将堆顶的最大数取出，这个过程持续到剩余数只有一个时结束。

最大堆：
-最大堆中的最大元素值出现在根结点（堆顶）
-堆中每个父节点的元素值都大于等于其孩子结点

```

#### (1) 使用heapq 库函数

````python
import heapq
def heap_sort(nums):
    res = []
    heapq.heapify(nums)
    while nums:
        item = heapq.heappop(nums)
        res.append(item)
    return res
````

````python
nums = [4,1,3,5,6,3,2,8]
heap_sort(nums)
````

#### （2） 自建堆

````python
def heapify(arr, index):
    left = 2*index+1
    right = 2*index+2
    max_index = index
    if (left < len(arr)) and (arr[left] > arr[max_index]):
        max_index = left
    if (right < len(arr)) and (arr[right] > arr[max_index]):
        max_index = right
    if max_index != index:
        arr[index], arr[max_index] = arr[max_index], arr[index]
        heapify(arr, max_index)

def heap_sort(nums):
    n = len(nums)
    res = []
    # 建堆
    for i in range(n,-1,-1):
        heapify(nums , i )

    # 从堆顶取元素
    for i in range(n-1, 0 ,-1):
        nums[i] , nums[0] = nums[0] , nums[i]
        heapify(nums, 0 ) 
````

````python
nums = [4,1,3,5,6,3,2,8]
heap_sort(nums)
nums
````
