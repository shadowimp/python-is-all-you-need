### 找数组中第k大的数
```
维护一个k大的最小堆，堆中是数组中最大的k个数。
第一个元素（nums[0])是堆顶元素
```

````python
import heapq
def kthLargestElement( k, nums):
    if len(nums) <k :
        return 
    nums_k = nums[0:k] 
    heapq.heapify(nums_k)
    for num in nums[k : len(nums) ]:
        item = max(num , heapq.heappop(nums_k))
        heapq.heappush(nums_k,item)
    return nums_k[0]
````

### 海量数据中取topk小的数
```
维护一个k大的大顶堆
新元素和堆顶元素比较，如果小于堆顶元素则入堆
```

````python
class Solution:
    def topK(self,nums,k):
        if not nums or len(nums)<k :
            return None 
        nums_k = nums[:k]
        self.build_heap(nums_k)
        for num in nums[k:]: 
            heap_top = nums_k[0]
            if num < heap_top:
                nums_k[0] = num
                self.heapify(nums_k,0)
        return nums_k
    
    def build_heap(self, arr):
        for i in range((len(arr)//2)-1, -1, -1):
            self.heapify(arr, i)
    
    def heapify(self, arr, index):
        # 大顶堆
        left = 2*index+1
        right = 2*index+2
        max_index = index
        if (left < len(arr)) and (arr[left] > arr[max_index]):
            max_index = left
        if (right < len(arr)) and (arr[right] > arr[max_index]):
            max_index = right
        if max_index != index:
            arr[index], arr[max_index] = arr[max_index], arr[index]
            self.heapify(arr, max_index)
````

````python
nums = [3,4,6,7,5,10,100,2]
k =2 
s = Solution()
s.topK(nums,k) 
````

### 调整数组顺序使奇数位于偶数前面

````python
def reOrderArray(nums):
    left , right =  0  , len(nums)-1
    while left < right:
        if nums[left] %2 == 0 and nums[right] %2 ==1 :
            nums[left] , nums[right] = nums[right] , nums[left]
            left += 1
            right -= 1
        if nums[left] % 2 == 1:
            left += 1
        if nums[right] %2 == 0:
            right -= 1
    return nums

print('a')
print(reOrderArray([1,2,3,4]) )
````

### 前序遍历(非递归）
```
从根节点开始，每次迭代弹出当前栈顶元素，并将其孩子节点压入栈中，
先压右孩子 -> 再压左孩子。
将出栈的元素依次append到 结果数组中。
```

````python
def pre_order(root):
    if not root:
        return None 
    res = []
    stack = []
    stack.append(root)
    while stack:
        node = stack.pop()
        res.append(node.val)
        if root.right:
            stack.append(root.right)
        if root.left:
            stack.append(root.left)
    return res
````

### 中序遍历(递归）

````python
def inorder(root):
    if not root:
        return []
    left = inorder(root.left)
    right = inorder(root.right)
    return left + [root.val] + right 
````

### 中序遍历(非递归）

````python
def inorder2(root):
    if not root:
        return None 
    res = []
    stack = []
    node = root
    while stack and node:
        if node:
            stack.append(node)
            node = node.left 
        else:
            node = stack.pop()
            res.append(node.val)
            node = node.left
    return res
    
    
    
````

### 中序遍历(非递归）

````python
def post_order(root):
    if not root:
        return None 
    node = root
    stack , out_stack = [], []
    while stack or node :
        if node :
            out_stack.append(node)
            stack.append(node)
            node = node.right
        else:
            node = stack
````

### 后序遍历（递归）

````python
def post_order(root):
    if not root :
        return []
    left = post_order(root.left)
    right = post_order(root.right)
    return left + right + [root.val]
````
