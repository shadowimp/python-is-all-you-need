````python
def problem1(s):
    length = len(s)
    if not s or s[0] == '0':
        return 0
    dp = [0]*(length+1)
    dp[0] = 1
    dp[1] = 1
    for i in range(2, length+1):
        if int(s[i-1]) >= 1 and int(s[i-1]) <= 9:
            dp[i] += dp[i-1]
        if int(s[i-2]) == 1 or (int(s[i-2]) == 2 and int(s[i-1]) >= 0 and int(s[i-1]) <= 6):
            dp[i] += dp[i-2]
    return dp[length]


s = input()
res = problem1(s)
print(res)
````

### 2.最长公共前缀

````python
def lcp(strs):
    if not strs: return ""
    s1 = min(strs)
    s2 = max(strs)
    for i,x in enumerate(s1):
        if x != s2[i]:
            return s2[:i]
    return s1
````
