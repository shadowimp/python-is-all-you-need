### 1. Two Sum

Given an array of integers, return indices of the two numbers such that they add up to a specific target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

Example:

Given nums = [2, 7, 11, 15], target = 9,

Because nums[0] + nums[1] = 2 + 7 = 9,
return [0, 1].

```
Two Sum: 一个数组中的两个数相加等于一个给定数，返回数组中两个数的下标。
输入： 一个数组，一个数字
输出： 两个数字
```

- 时间复杂度： O(n)
- 空间复杂度： O(n)

````python
def two_sum(nums,target):
    nums_dict = { }  #创建一个空字典,将target-num1作为key，num1的下标作为value。
    for i in range(len(nums)):
        if nums[i] not in nums_dict: #如果没有，则将 target-nums1 和 nums1的下标 存入字典nums_dict
            nums_dict[target-nums[i]]  = i
        else:       #遍历字典看其中是否有x，如果有说明两个数找到了
            return [nums_dict[nums[i]],i]
````

### 20.Valid Parentheses  
Given a string containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

1.  if 元素为左： 入栈
2. else : 与栈中弹出的元素比较，能不能构成一对。
3. 最后栈中元素为空， 返回True 
- 时间复杂度： O(n)
- 空间复杂度： O(n)

````python
def valid_parent(s):
    stack = []
    dic = {'(':')','[':']','{':'}'}
    
    for c in s:
        if c in dic:
            stack.append(c)
        else:
            if not stack or dic[stack.pop()] != c:
                return False 
    return not stack 
````

### 21. Merge Two Sorted Lists  合并两个有序链表

````python
class LinkNode:
    def __init__(self,x):
        self.val = x
        self.next = None

def merge_list(l1,l2):
    s = l = LinkNode(0) #初始化一个头节点， 最后返回头节点的next
    while l1 and l2 :
        if l1.val < l2.val:
            l.next = LinkNode(l1.val)
            l1 = l1.next 
        else:
            l.next = LinkNode(l2.val)
            l2 = l2.next 
        l = l.next 
    
    while l1:
        l.next = LinkNode(l1.val)
        l = l.next 
        l1 = l1.next
    while l2:
        l.next = LinkNode(l2.val)
        l = l.next 
        l2 = l2.next 
    return s.next 
````

### 53. Maximum Subarray连续子数组的最大和
```
Given an integer array nums, find the contiguous subarray (containing at least one number) which has the largest sum and return its sum.

Input: [-2,1,-3,4,-1,2,1,-5,4]

Output: 6

Explanation: [4,-1,2,1] has the largest sum = 6.
```

````python
def max_subarray(nums):
    if not nums:
        return None 
    max_sum = current_sum = nums[0]
    for num in nums[1:]: 
        current_sum = max(current_sum+num,num)  #连续和加当前元素，与当前元素比较
        max_sum = max(current_sum,max_sum)
    return max_sum   
````

### 70. Climbing Stairs
You are climbing a stair case. It takes n steps to reach to the top.

Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?

````python
def climb_stairs(n):
    if n == 1 :
        return 1 
    if n == 2 :
        return 2 
    
    steps = [ 0 for i in range(n) ]
    steps[0] , steps[1] = 1,2
    for i in range(2,n):
        steps[i] = steps[i-1] + steps[i-2]
    return steps[-1] 
````

````python
def symmetric_tree(root):
    if root is None :
        return True 
    else :
        return fun(root.left,root.right)
def fun(left,right):
    """
    判断两边是否是镜像，必须传入两个参数
    """
    if left is None and right is None :
        return True 
    if left is None or right is None:
        return False
    if left.val != right.val:
        return False
    else:
        flag1 = fun(left.left,right.right)
        flag2 = fun(left.right,right.left)
        return flag1 and flag2
````

### 155. Min Stack
Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

````python
class MinStack:
    def __init__(self):
        self.s = []
    
    def push(self,x):
        if not self.s or x<self.s[-1][1]:  # 栈中的每个元素存储两个值，一个是入栈的元素值，一个是当前栈中的最小值
            self.s.append((x,x))
        else:
            self.s.append((x,self.s[-1][1]))
            
    def pop(self):
        return self.s.pop()
    
    def top(self):  #栈是后进先出，所以栈顶元素就是我们构建的列表中最后的元素
        return self.s[-1][0]
    
    def getMin(self):
        return self.s[-1][1]
    
````

### 169. Majority Element
找到数组中出现次数大于n/2的数组 ， 数组非空且majority element一定存在

````python
from collections import Counter
def majority_element(nums):
    count = Counter(nums)
    for i , j in count.items() :
        if j > len(nums)/2:
            return i
````

### 234. Palindrome Linked List
判断一个链表是否是回文，用 O(1) 空间复杂度解决
```
Example 1:
Input: 1->2
Output: false

Example 2:
Input: 1->2->2->1
Output: true

```

````python
def palindrome_list(head):
    """
    快慢指针 + 反转链表
    """
    slow = fast = head 
    
    #find the middle 
    while fast and fast.next:
        fast = fast.next.next 
        slow = slow.next 
    
    # reverse the second half 
    new = None 
    while slow :
        temp = slow 
        slow = slow.next
        temp.next = new
        new = temp 
    # compare the two half 
    while head :
        if head.val != new.val :
            return False 
        head = head.next 
        new = new.next
    return True 
    
````

### 617. Merge Two Binary Trees
将两颗二叉树合并， 如果同一个位置，两颗树都有值，则两值相加， 否则用原来不空的值代替
```
Input: 
	Tree 1                     Tree 2                  
          1                         2                             
         / \                       / \                            
        3   2                     1   3                        
       /                           \   \                      
      5                             4   7                  
Output: 
Merged tree:
	     3
	    / \
	   4   5
	  / \   \ 
	 5   4   7
```

````python
def merge_tree(t1, t2):
    # 将树t2上的值加到树t1的值上 ， 最后返回t1
    if not t2 : 
        return t1
    if not t1 :
        return t2
    
    t1.val += t2.val 
    t1.left = merge_tree(t1.left, t2.left)
    t2.right = merge_tree(t1.right, t2.right)
    
    return t1 
    
````
