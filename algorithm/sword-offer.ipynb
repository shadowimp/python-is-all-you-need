### 使数组的奇数在前，偶数在后。

````python
def odd_even(nums):
    i , j = 0 , len(nums)-1
    while i < j : 
        if nums[i] % 2 == 0 and nums[j] % 2 == 1:
            nums[i] , nums[j] = nums[j] , nums[i]
            i = i + 1 
            j = j - 1 
        elif nums[i] % 2 == 0 :
            j = j - 1 
        elif nums[j] % 2 == 1 :
            i = i+1 
        else:
            i = i+1
            j = j -1 
    
    return nums 
````

### 16 反转链表

````python
class LinkNode:
    def __init__(self,val):
        self.val = val
        self.next = None
def linknode_reverse(head):
    if not head :
        return None
    new_head = None # 这个None最后是 最后一个节点的next
    while head :
        temp = head
        head = head.next
        temp.next = new_head
        new_head = temp
    return new_head
````

````python
l1 = LinkNode(3)
l1.next = LinkNode(4)
l1.next.next = LinkNode(6) 
````

````python
l = linknode_reverse(l1)
````

### 17 合并两个排序的链表

````python
def merge_linklist(l1, l2):
    if not l1 :
        return l2 
    if not l2 :
        return l1
    new_head = l3 = LinkNode(0)
    while l1 and l2:
        if l1.val < l2.val:
            l3.next = LinkNode(l1.val)
            l3 = l3.next
            l1 = l1.next
        else:
            l3.next = LinkNode(l2.val)
            l3 = l3.next
            l2 = l2.next
    if l1:
        l3.next = l1
    if  l2:
        l3.next = l2
    return new_head.next
````

### 旋转数组的最小元素
把一个数组最开始的若干个元素搬到数组的末尾，我们称之为数组的旋转。 输入一个非减排序的数组的一个旋转，输出旋转数组的最小元素。 例如数组{3,4,5,1,2}为{1,2,3,4,5}的一个旋转，该数组的最小值为1。 NOTE：给出的所有元素都大于0，若数组大小为0，请返回0。

````python
def min_rotate(nums):
    if not nums:
        return 
    if nums[0] < nums[-1] :
        return nums[0]
    left , right = 0 , len(nums) -1 
    while left < right :
        mid = (left + right) //2 
        if nums[mid] > nums[left]:
            right = mid 
        else:
            left = mid
````

0 ，1, 1, 2, 3, 5, 8, 13 。。。 斐波那契数列，现在要求输入一个整数n，请你输出斐波那契数列的第n项（从0开始，第0项为0）。

````python
def fibonacci(n):
    # f(n) = f(n-1) + f(n-2)
    if n == 0:
        return 0 
    if n == 1 :
        return 1 
    n1, n2 = 0,1 
    # print(n1,n2)
    for i in range(1,n):
        temp  = n2 
        n2 = n1+n2 
        n1 = temp
    return n2 
````

### 求字符串的逆序数

````python
def inverse_num(string):
    ans = 0
    for i in range(len(string)):
        for j in range(i):
            if string[j] > string[i]:
                ans += 1
    return ans
````
